# **Competition Code**

## 1. Preprocessing Steps

```python
import pandas as pd
import json

# Turn posts JSON into df
with open("./dm-lab-2-private-competition/final_posts.json", "r") as f:             
    data = json.load(f)

posts = []
for entry in data:
    post = entry["root"]["_source"]["post"]
    posts.append({
        "id": post["post_id"],
        "text": post["text"],
    })

# Convert to dataframe
posts_df = pd.DataFrame(posts)
```

```python
# Create and finalize the train and test dfs

df = pd.read_csv("./dm-lab-2-private-competition/data_identification.csv")
y = pd.read_csv("./dm-lab-2-private-competition/emotion.csv")    

# Add the text feature
df = df.merge(posts_df, on="id", how="left")

# Split into train/validation and test dataframes
train_df = df[df['split'] == 'train'].reset_index(drop=True)
test_df = df[df['split'] == 'test'].reset_index(drop=True)

print("Train/Validation size:", train_df.shape)
print("Test size:", test_df.shape)
display(train_df.head(5))
display(test_df.head(5))
```

```python
# Drop the "id" column in the df because it shouldn't be used for training
y = y.drop(columns=["id"])

# Drop the "split" column in the df because it shouldn't be used for training and isn't needed for anything else
test_df = test_df.drop(columns=["split"])
train_df = train_df.drop(columns=["split"])
display(y.head(5))
```

## 2. Feature Engineering Steps

```python
### Track the number of exclamation marks (joy, surprise), question marks (surprise), and capitalized words (disgust, anger)
train_df['exclamation_mark_count'] = train_df['text'].str.count('!')
train_df['question_mark_count'] = train_df['text'].str.count('\?')
train_df['capitalized_word_count'] = train_df['text'].apply(lambda x: sum(1 for w in x.split() if w.isupper()))

test_df['exclamation_mark_count'] = test_df['text'].str.count('!')
test_df['question_mark_count'] = test_df['text'].str.count('\?')
test_df['capitalized_word_count'] = test_df['text'].apply(lambda x: sum(1 for w in x.split() if w.isupper()))

display(train_df.head(5))
display(test_df.head(5))
```

## 3. Model Implementation Steps

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    train_df[['text', 'exclamation_mark_count', 'question_mark_count', 'capitalized_word_count']],
    y,
    test_size=0.2,
    random_state=0
)
display(X_train.head(5))
display(y_train.head(5))
display(X_val.head(5))
display(y_val.head(5))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, f1_score
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

numerical_columns = ['exclamation_mark_count', 'question_mark_count', 'capitalized_word_count']

# Need to standardize
preprocessor = ColumnTransformer(
    transformers=[
        ('tfidf', TfidfVectorizer(), 'text'),
        ('numerical', StandardScaler(), numerical_columns)
    ]
)

model = Pipeline([
    ('preprocess', preprocessor),
    ('clf', LogisticRegression(max_iter=500, class_weight='balanced'))
])

model.fit(X_train, y_train)

y_val_pred = model.predict(X_val) 


val_acc = accuracy_score(y_val, y_val_pred)
val_f1w = f1_score(y_val, y_val_pred, average='weighted')

print(y_val_pred)
print(y_val)
print(f"[Base - Validation] Acc: {val_acc:.4f} | F1-weighted: {val_f1w:.4f}")
```

```python
y_test_pred = model.predict(test_df[['text', 'exclamation_mark_count', 'question_mark_count', 'capitalized_word_count']]) 

# Create submission dataframe
submission_df = pd.DataFrame({
    "id": test_df['id'], 
    "emotion": y_test_pred
})

# Save to CSV
submission_df.to_csv("submission.csv", index=False)

print("Done writing the submission file")
```
